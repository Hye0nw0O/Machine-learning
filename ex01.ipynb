```python
# 키 195 -> 극단치. 데이터 전처리를 통해 극단치를 처리해야 함.
height = [155.3, 157.5, 156.5, 163.9, 169.3, 170.5, 175.1, 173.8, 177.7, 182.6, 180.7, 186.7, 189.9, 189.2]
foot_size = [220, 230, 235, 240, 245, 250, 260, 265, 270, 275, 280, 285, 290, 300]
```

```python
plt.scatter(foot_size, height)
plt.xlabel('Foot size')
plt.ylabel('Height')
plt.show()
```

```python
x = np.array(foot_size)
y = np.array(height)
```

```python
a = np.sum((y-np.mean(y))*(x-np.mean(x)))
a = a/np.sum((x-np.mean(x))**2)
b = np.mean(y) - a*np.mean(x)
print('a(기울기): ', a, 'b(절편): ', b)
```

```python
line_x = np.arange(min(x), max(x), 1)
line_y = a*line_x + b
```

```python
plt.scatter(line_x, line_y)
plt.scatter(x, y)
plt.xlabel('Foot size')
plt.ylabel('Height')
plt.show()
```
